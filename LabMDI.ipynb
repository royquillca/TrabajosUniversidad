```python
#IMOPRTACION DE LIBRERÍAS
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#ACELERACION DEL PUNTO P: COMPONENTE X
t = np.linspace(0, 0.2446, 100)
tita = np.linspace(np.pi/6, np.pi/2, 100)
beta = np.linspace(0, -1.9957, 100)
def theta (t):
  return (np.pi/6)+4*t+1.15*(t**2)
def bheta(t):
  return -8*t-0.65*(t**2)
def acelAx(t):
  return -552*np.sin(theta(t))+276*np.cos(bheta(t))-240*((2.3*t+4)**2)*np.cos(theta(t))+120*((2.3*t+4)**2)*np.sin(bheta(t))-156*np.cos(bheta(t))+120*((8+1.3*t)**2)*np.sin(bheta(t))-(7680+5664*t+717.6*t**2 )*np.sin(bheta(t))

plt.plot(t, acelAx(t), label="Aceleración Abs. Eje X")
plt.xlabel("Tiempo[s]")
plt.ylabel("Aceleración [m/s^2]")
plt.title("Aceleración del punto P Componente X")
plt.grid()
plt.legend(loc=1)
```

```python
acelAx(0)
```

```python
#ACELERACION DEL PUNTO P: COMPONENTE Y
#cambiar t= np.linspace(0, tiempoCuando_Theta=90, 100)
t = np.linspace(0, 0.2446, 100)
tita = np.linspace(np.pi/6, np.pi/2, 100)
beta = np.linspace(0, -1.9957, 100)
def theta (t):
  return (np.pi/6)+4*t+1.15*(t**2) # Funcion theta(t) 
def bheta(t):
  return -8*t-0.65*(t**2) #Funcion bheta(t) del sistema acoplado al brazo
def acelAy(t):
  #Modificar la funcion Aceleratiion(t) en funcion del tiempo, theta y/o bheta segun sea necesario
  return 552*np.cos(theta(t))-276*np.sin(bheta(t))-((2.3*t+4)**2)*(240*np.sin(theta(t))-120*np.cos(bheta(t)))+156*np.sin(bheta(t))+120*((8+1.3*t)**2)*np.cos(bheta(t))-(7680+5664*t+717.6*(t**2) ) *np.cos(bheta(t))
plt.plot(t, acelAy(t), 'r',label="Aceleración Abs. Eje y")
plt.xlabel("Tiempo[s]")
plt.ylabel("Aceleración [m/s^2]")
plt.title("Aceleración del punto P - Componente Y")
plt.grid()
plt.legend(loc=1)
```

```python
acelAy(0)
```

```python
#Aceleracion total ABSOLUTA
def acelAbs(t):
  return np.sqrt(acelAx(t)**2+acelAy(t)**2)
plt.plot(t, acelAbs(t), label="Aceleración Abs. de P")
plt.xlabel("Tiempo[s]")
plt.ylabel("Aceleración [m/s^2]")
plt.title("Aceleración Absoluta del punto P")
plt.grid()
plt.legend(loc=1)
```
